```python
import networkx as nx
from networkx.algorithms.community.centrality import girvan_newman
from networkx.algorithms.community.kclique import k_clique_communities
from networkx.algorithms.community import greedy_modularity_communities
import matplotlib.pyplot as plt
```

```python
L1 = nx.read_edgelist('./data/layer1.txt')
L2 = nx.read_edgelist('./data/layer1.txt')
L3 = nx.read_edgelist('./data/layer1.txt')
L4 = nx.read_edgelist('./data/layer1.txt')
```

```python
print("*****Layer1*****")
communities_layer1 = list(greedy_modularity_communities(L1))
print([list(x) for x in communities_layer1])
print("*****Layer2*****")
communities_layer2 = list(greedy_modularity_communities(L2))
print([list(x) for x in communities_layer2])
print("*****Layer3*****")
communities_layer3 = list(greedy_modularity_communities(L3))
print([list(x) for x in communities_layer3])
print("*****Layer4*****")
communities_layer4 = list(greedy_modularity_communities(L4))
print([list(x) for x in communities_layer4])
```

```python
%matplotlib inline
```

```python
pos = nx.spring_layout(L1)
colors = ["indigo", "orange", "red", "blue", "green", "yellow"]
for i in range(len(communities_layer1)):
    graph = communities_layer1[i]
    node_list = [node for node in graph]
    nx.draw(L1, pos, nodelist=node_list, node_color=colors[i%6], node_size=50, alpha=0.8)
```
